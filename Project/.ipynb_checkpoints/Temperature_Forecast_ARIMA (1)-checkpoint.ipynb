```python
import pandas as pd
import numpy as np
```

#Read Data

```python
df=pd.read_csv('AAPL_data.csv',index_col='date',parse_dates=True)
df=df.dropna()
print('Shape of data',df.shape)
df.head()
```

#Plot Your Data

```python
df['close'].plot(figsize=(12,5))
```

#Check For Stationarity

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(dataset):
    dftest = adfuller(dataset, autolag = 'AIC')
    print("1. ADF : ",dftest[0])
    print("2. P-Value : ", dftest[1])
    print("3. Num Of Lags : ", dftest[2])
    print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
    print("5. Critical Values :")
    for key, val in dftest[4].items():
          print("\t",key, ": ", val)
```

```python
adf_test(df['close'])
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

#Split Data into Training and Testing

```python
print(df.shape)
train=df.iloc[:-30]
test=df.iloc[-30:]
print(train.shape,test.shape)
print(test.iloc[0],test.iloc[-1])
```

## Train the Model

```python
train['close']
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(train['close'],order=(6,2,1))
model=model.fit()
model.summary()
```

#Make Predictions on Test Set

```python
start = len(train)
end = len(train) + len(test) - 1

# If the predicted values don't have date values as index, you will have to uncomment the following two commented lines to plot a graph
# index_future_dates = pd.date_range(start='2018-12-01', end='2018-12-30')
pred = model.predict(start=start, end=end, typ='levels').rename('ARIMA predictions')
pred.index = test.index  # Set the index of the predicted values to match the index of your test data

# pred.index = index_future_dates  # Uncomment this line if the predicted values don't have date values as index

pred.plot(legend=True)
test['close'].plot(legend=True)
```

```python
pred.plot(legend='ARIMA Predictions')
test['close'].plot(legend=True)
```

```python
test['close'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(pred,test['close']))
print(rmse)
```
